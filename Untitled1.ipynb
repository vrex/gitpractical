```python
import numpy as np

# Dictionary to store embeddings
embeddings_dict = {}

# Open GloVe file
with open("glove.6B.100d.txt", 'r', encoding='utf-8') as f:
    
    for line in f:
        
        values = line.split()
        
        word = values[0]
        
        vector = np.asarray(values[1:], dtype='float32')
        
        embeddings_dict[word] = vector

print("Total words loaded:", len(embeddings_dict))
```

```python
embeddings_dict["king"]
```

```python
from numpy.linalg import norm

def cosine_similarity(word1, word2):
    
    vec1 = embeddings_dict[word1]
    vec2 = embeddings_dict[word2]
    
    similarity = np.dot(vec1, vec2) / (norm(vec1) * norm(vec2))
    
    return similarity

# Example
print(cosine_similarity("king", "queen"))
print(cosine_similarity("cat", "dog"))
```

```python
def most_similar(word, top_n=5):
    
    similarities = {}
    
    target_vector = embeddings_dict[word]
    
    for other_word in embeddings_dict:
        
        if other_word == word:
            continue
        
        other_vector = embeddings_dict[other_word]
        
        similarity = np.dot(target_vector, other_vector) / (
            norm(target_vector) * norm(other_vector)
        )
        
        similarities[other_word] = similarity
    
    # Sort words
    sorted_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_words[:top_n]

# Example
print(most_similar("computer"))
```

```python
def word_analogy(word1, word2, word3, top_n=5):
    
    # vector(word2) - vector(word1) + vector(word3)
    result_vector = (
        embeddings_dict[word2]
        - embeddings_dict[word1]
        + embeddings_dict[word3]
    )
    
    similarities = {}
    
    for word in embeddings_dict:
        
        similarity = np.dot(result_vector, embeddings_dict[word]) / (
            norm(result_vector) * norm(embeddings_dict[word])
        )
        
        similarities[word] = similarity
    
    sorted_words = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_words[:top_n]

# Example
print(word_analogy("man", "king", "woman"))
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
words = [
    "king", "queen", "man", "woman",
    "dog", "cat", "lion", "tiger",
    "apple", "banana", "orange",
    "car", "bus", "train"
]

word_vectors = np.array([embeddings_dict[word] for word in words])
```

```python
pca = PCA(n_components=2)

reduced_vectors = pca.fit_transform(word_vectors)

plt.figure(figsize=(10,8))

for i, word in enumerate(words):
    
    x = reduced_vectors[i][0]
    y = reduced_vectors[i][1]
    
    plt.scatter(x, y)
    plt.text(x+0.01, y+0.01, word)

plt.title("PCA Visualization of Word Embeddings")
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# Convert to float64
word_vectors = np.array(word_vectors, dtype=np.float64)

# Create TSNE model
tsne = TSNE(
    n_components=2,
    random_state=42,
    perplexity=5,
    method='exact'
)

# Transform vectors
tsne_vectors = tsne.fit_transform(word_vectors)

# Plot
plt.figure(figsize=(10,8))

for i, word in enumerate(words):

    x = tsne_vectors[i][0]
    y = tsne_vectors[i][1]

    plt.scatter(x, y)
    plt.text(x+0.5, y+0.5, word)

plt.title("t-SNE Visualization")
plt.show()
```
